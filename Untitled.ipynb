```python
from sklearn.decomposition import PCA
from dim_methods import get_varimax_loadings_standard as varimax

class PCA_dm:
    def __init__(self, data, n_components = None, time_first = True):
        self.dataset = data
        self.n_components = n_components
        
        if len(self.dataset) > 3:
            self.dataset = self.dataset.reshape(self.dataset.shape[0], -1)
            
        if time_first:
            self.dataset = self.dataset.transpose()
        
    def perform_pca(self):
        self.pca = PCA(n_components = self.n_components)
        self.pca.fit(self.dataset)
```

```python
latent_data = np.load("data/outputs.npy")
dm_output = PCA_dm(latent_data.reshape(900, 32))
```

```python
dm_output.perform_pca()
```

```python
for i in range(4):
    plt.plot(volc_activity_np[:900])
    plt.plot(dm_output.pca.components_[i, :])
    plt.show()
```
